**Adding new column with assign and insert**

1. assign()
When to use assign?
 - Add multiple columns in a single line of code
 - When you need to overwrite the values of an existing columns (best practise)

 It returns a new object (a copy) with all original columns in addition to the new ones

```python
import numpy as np
```

```python
score1 = np.random.randint(1,100,size=1000)
score2 = np.random.randint(1,100,size=1000)
```

```python
series1 = pd.Series(score1, index=np.arange(0,1000))
series2 = pd.Series(score2, index=np.arange(0,1000))
```

```python
df_exams = df_exams.assign(score1=series1, score2=series2)
```

```python
df_exams.head()
```

 2. insert()

 Inserts a new column at a specific position or index 

```python
df_exams.insert(1, 'test', series1)
```

```python
df_exams.head()
```

In assign we have reassign to the variable as a copy is created but in insert no copy is created.
