For multiple CSVs

```python
root = "https://www.football-data.co.uk/mmz4281/"
```

```python
leagues = ["E0", "E1", "E2", "E3", "EC"]
frames = []

for league in leagues:
    csv_path = root + "2122/" + league + ".csv"
    df = pd.read_csv(csv_path, encoding='unicode_escape')
    frames.append(df)
```

For multiple seasons

```python
leagues = ['E0', "E2", "E3"]
frames = []

for league in leagues:
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + league)
        df.insert(1,'season', season)
        frames.append(df)
```

# Organise all data in a dictionary

```python
dict_countries={
    'Spanish La Liga': 'SP1', 'Spanish Segunda Division': 'SP2', 'German Bundesliga': 'D1',
    'English Premier League': 'E0', 'English League 1': 'E2', 'English League 2': 'E3'
}
```

```python
dict_countries['Spanish La Liga']
```

```python
dict_historical_data = {}

for league in dict_countries:
    frames = []
    for season in range(15,21):
        df = pd.read_csv(root + str(season) + str(season+1) + '/' + dict_countries[league])
        df.insert(1,'season', season)
        frames.append(df)
    df_concat = pd.concat(frames)
    dict_historical_data[league] = df_concat
```
